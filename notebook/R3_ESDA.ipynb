```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
pre_covid_daily_train_demand = pd.read_csv('../data/curated/train_demand/daily_trains_demand_pre_covid.csv')
post_covid_daily_train_demand = pd.read_csv('../data/curated/train_demand/daily_trains_demand_post_covid.csv')
```

```python
pre_covid_daily_train_demand[['Passenger_Boardings', 'Passenger_Alightings', 'Total_Demand', 'log_Total_Demand', 'log_Passenger_Boardings', 'log_Passenger_Alightings']].describe()
```

```python
post_covid_daily_train_demand[['Passenger_Boardings', 'Passenger_Alightings', 'Total_Demand', 'log_Total_Demand', 'log_Passenger_Boardings', 'log_Passenger_Alightings']].describe()
```

```python
import matplotlib.pyplot as plt

plot_list = ['log_Total_Demand', 'log_Passenger_Boardings', 'log_Passenger_Alightings']
data_list = [pre_covid_daily_train_demand, post_covid_daily_train_demand]
data_names = ['Pre-Covid', 'Post-Covid']

# Create subplots
fig, axes = plt.subplots(len(data_list), len(plot_list), figsize=(15, 10))

# Plot data
for i in range(len(data_list)):
    for j, plot in enumerate(plot_list):
        ax = axes[i, j]
        data_list[i][plot].hist(ax=ax)
        data_name = data_names[i]
        ax.set_title(f'{data_name} {plot}')

# Adjust layout to prevent overlap
plt.tight_layout()
plt.show()
```
